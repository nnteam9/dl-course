Как вы наверно знаете, питон - *интерпретируемый* ЯП. Что позволяет пользоваться такими фишками, как интерактивный режим питона и юпитер, где вы можете независимо прогонять код, который будет иметь общие переменные в оперативной памяти, но не будет требовать наличие всего кода во время запуска.

По сути программа на юпитере это башня лего, к которой вы сверху пристраиваете кирпичики с кодом. Причем новые могут иметь за основу предыдущие, а также вы можете менять порядок в котором добавляете блоки.

### Пример

```python
a = 5
```

```python
a = 15
```

```python
print(a)
```

В первой ячейке вы создаете переменную "a", типа int, которая будет жить среди глобальных переменных в вашей программе. Во второй клетке - вы меняете её значение на 15. Но никто не мешает изменить порядок запуска команд и получить в последней ячейке 5, а не 15 в качестве вывода (также как и NameError: name 'a' is not defined).

Это свойство питона и юпитера, на самом деле и определило популярность этих технологий в Data Science. Вам не нужно каждый раз загружать данные, запускать отладчик, чтобы попробовать пару функций в библиотке, не нужно создавать новый файл с кодом, для быстрого эксперимента.

Главный принцип написания кода - идти от простого к сложному, постепенно, проверяя себя на каждом шагу (запуская код в ячейке).

Питон является объектно-ориентированным языком, и поддерживает классы, наследования, методы

```python
# давайте сразу поработаем с данными, для этого нам понадобится библиотека для работы
# с файловой системой (встроенная)
import os
```

```python
# чтобы создать класс используйте ключевое слово class
class Dataset:  # кстати для названия классов принято использовать PascalCase
    # чтобы вызывать конструктор, питон использует метод __init__ (дандер (double-underscore) инит)
    # поэтому для того, чтобы задать конструктор необходимо определить этот метод
    def __init__(self, directory):  # первый аргумент (self) будет использоваться для передачи методу 
                                    # инстанса самого класса (который создастся автоматически перед вызовом)
                                    # но изначально он будет пустой, поэтому в конструкторе мы опишем
                                    # все переменные инстанса которые нас интересуют
        # давайте получим список файлов в директории и присвоим его соотвествующему полю в инстансе класса
        self.files_paths = os.listdir(directory)  # listdir вернёт список всех файлов (±ls в unix)
        self.directory = directory
    
    # а теперь создадим обычный метод
    def get_images(self):
        # в питоне есть замечательный способ создавать списки - list comprehension
        # синтаксис: [func(x) for x in iterable if func2(x)], - эквивалентно:
        # result = []
        # for x in iterable:
        #     if func2(x) == True:
        #          result.append(func(x))
        return [
            os.path.join(self.directory, x)  # метод path.join добавляет "/" между частями пути
            for x in self.files_paths  # итерируемся по названиям файлов
            if (x.endswith('.jpg') or x.endswith('.jpeg') or x.endswith('.png')) \
                                                                                   # \ backslash позволяет
                and not x.startswith('.')                                          # продолжить код на другой
        ]                                                                          # строке
```

```python
dataset = Dataset('..')  # теперь мы можем создать инстанс нашего класса используя круглые скобки
dataset.get_images()  # и вызвать наш метод
```

```python
# для deep понимания
dataset = Dataset.__new__(Dataset)
Dataset.__init__(dataset, '..')
dataset.get_images()
```
