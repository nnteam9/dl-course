## Основы работы

Итак, надеюсь вы читаете этот текст уже из своего юпитера.  
В юпитер ноубуке существует два типа редактирования - редактирование клеток и текста. Для редактирования клетки - кликните на неё, но не по полю с текстом. Теперь вы можете делать различные манипуляции с клеткой - вы можете вставить пустую клетку под ней (клавиша B), скопировать (С), вставить (V) и изменить её тип, нажав M. (более полный лист хот-кеев вы найдете под кнопкой "Help" выше).   
Выделите клетку ниже и измените её тип с Markdown на Code, нажав Y.

''.join([chr(ord(x) ^ 42) for x in 'bKZZS\nzS^BEDCDM'])

Чтобы выполнить код в клетке, нажмите Ctrl+Enter или Shift+Enter. Также это работает с набором текста в Markdown для компиляции в html код. Вывод результата прогона клетки происходит под самой клеткой.

Как вы наверно знаете, питон - *интерпретируемый* ЯП. Что позволяет пользоваться такими фишками, как интерактивный режим питона и юпитер, где вы можете независимо прогонять код, который будет иметь общие переменные в оперативной памяти, но не будет требовать наличие всего кода во время запуска.

По сути программа на юпитере это башня лего, к которой вы сверху пристраиваете кирпичики с кодом. Причем новые могут иметь за основу предыдущие, а также вы можете менять порядок в котором добавляете блоки.

### Пример

```python
a = 5
```

```python
a = 15
```

```python
print(a)
```

В первой ячейке вы создаете переменную "a", типа int, которая будет жить среди глобальных переменных в вашей программе. Во второй клетке - вы меняете её значение на 15. Но никто не мешает изменить порядок запуска команд и получить в последней ячейке 5, а не 15 в качестве вывода (также как и NameError: name 'a' is not defined).

Это свойство питона и юпитера, на самом деле и определило популярность этих технологий в Data Science. Вам не нужно каждый раз загружать данные, запускать отладчик, чтобы попробовать пару функций в библиотке, не нужно создавать новый файл с кодом, для быстрого эксперимента.

Главный принцип написания кода - идти от простого к сложному, постепенно, проверяя себя на каждом шагу (запуская код в ячейке).

1. Синтаксис

```python
# это комментарий
# а так задается функция:
def some_function_name(first_argument, second_argument_with_default_value=42):
    # код внутри функции пишется с отступом в один tab (4 пробела)
    result = first_argument ** 2 + second_argument_with_default_value  # возведение в степень - **
    # if else выглядит так
    if result < 100:
        return str(result)  # конвертация в строковую переменную
    else:
        return '0'  # вы можете использовать для строк как " так и '
```

```python
# вызов функции производится как обычно, за исключением опционального значения:
a = some_function_name(9, 10)
b = some_function_name(9, second_argument_with_default_value=10)
c = some_function_name(9)
```

```python
print(a, b, c)  # в функцию print можно передавать сразу несколько аргументов
```

```python
# Массивы в питоне "представлены" классом list и tuple.
# list - динамический массив, в который можно положить значения, изменить, удалить или добавить.
list_wth_some_values = [1, "1", 1., some_function_name, None]  # list можно создать используя []
# внутри листа могут храниться любые объекты, даже функции
list_wth_some_values[2](9)  # по умолчанию, когда вы запускаете функцию, возвращающую значение
# в конце клетки и не передаете его никуда - оно выводится в вывод клетки
# (на самом деле функции всегда что-то возвращают, просто None не обрабатывается юпитером)
```

```python
# кстати 1. эквивалентно 1.0 или float(1) и имеет тип float
# здесь, в чистом питоне нам не имеет значения fp16, 24 или 32. Для точных вычислений есть специальные библиотки,
# о которых позднее. Кстати, int - "бесконечный", имеет максимальный размер равный машинному слову
10**1015
```

```python
# Циклы можно создавать используя привычные for/while операторы
for i in range(10):  # range - создает итератор, - объект на лету создающий значения. Здесь - целые от 0 до 9 включительно
    print(i, end=' ')  # аргумент end позволяет закончить вывод тем символом, который укажете
# посмотреть подсказку / документацию к функции - shift+tab
```
